```python
from nba_api.stats.endpoints import leaguegamelog
from pathlib import Path
import pandas as pd
import time

DATA_RAW = Path("../data_raw")
DATA_RAW.mkdir(exist_ok=True)

def download_league_gamelog_with_retry(
    season="2025-26",
    season_type="Regular Season",
    out_path=DATA_RAW / "league_gamelog_2025_26.csv",
    max_tries=8,
    sleep_seconds=4
):
    last_err = None

    for attempt in range(1, max_tries + 1):
        try:
            print(f"üì• LeagueGameLog deneme {attempt}/{max_tries}...")

            lg = leaguegamelog.LeagueGameLog(
                season=season,
                season_type_all_star=season_type
            )

            df = lg.get_data_frames()[0]

            out_path.parent.mkdir(parents=True, exist_ok=True)
            df.to_csv(out_path, index=False)

            print(f"‚úÖ Kaydedildi: {out_path.resolve()}")
            print(f"Toplam satƒ±r (team-level): {len(df)}")
            return df

        except Exception as e:
            last_err = e
            print(f"‚ö†Ô∏è Hata: {type(e).__name__}: {e}")
            print(f"‚è≥ {sleep_seconds} sn bekleniyor...\n")
            time.sleep(sleep_seconds)

            # exponential backoff
            sleep_seconds = min(sleep_seconds * 2, 30)

    raise RuntimeError(f"‚ùå LeagueGameLog indirilemedi. Son hata: {last_err}")


# √áALI≈ûTIR
df = download_league_gamelog_with_retry()
```
