```python
import pandas as pd
from pathlib import Path

DATA_RAW = Path("../data_raw")
DATA_RAW.mkdir(exist_ok=True)

print("Kayıt yolu:", DATA_RAW.resolve())
```

```python
# 2025–26 oyuncu istatistikleri (NBAstuffer)
url_player = "https://www.nbastuffer.com/2025-2026-nba-player-stats/"

# Sayfadaki ilk tabloyu al
df_players = pd.read_html(url_player)[0]

# İlk 5 satırı göster
display(df_players.head())

# Kaç oyuncu var?
print("Oyuncu sayısı:", len(df_players))

# CSV olarak kaydet
output_path = DATA_RAW / "nbastuffer_2025_2026_player_stats_raw.csv"
df_players.to_csv(output_path, index=False)
print("Kaydedildi:", output_path)
```

```python
# 2025–26 takım istatistikleri (NBAstuffer)
url_team = "https://www.nbastuffer.com/2025-2026-nba-team-stats/"

df_teams = pd.read_html(url_team)[0]

display(df_teams.head())
print("Takım sayısı:", len(df_teams))

output_path = DATA_RAW / "nbastuffer_2025_2026_team_stats_raw.csv"
df_teams.to_csv(output_path, index=False)
print("Kaydedildi:", output_path)
```

```python
DATA_RAW = Path("../data_raw")
DATA_RAW.mkdir(exist_ok=True)

urls = {
    "referee_stats": "https://www.nbastuffer.com/2025-2026-nba-referee-stats/",
    "rest_days_stats": "https://www.nbastuffer.com/2025-2026-nba-rest-days-stats/",
    "schedule_rest_days": "https://www.nbastuffer.com/2025-2026-nba-schedule-rest-days-analysis/"
}

for name, url in urls.items():
    df = pd.read_html(url)[0]
    df.to_csv(DATA_RAW / f"nbastuffer_2025_2026_{name}.csv", index=False)
    print(f"✓ Kaydedildi: nbastuffer_2025_2026_{name}.csv")
```

```python
from nba_api.stats.endpoints import playergamelog
from pathlib import Path

DATA_RAW = Path("../data_raw")
DATA_RAW.mkdir(exist_ok=True)

# Örnek oyuncu: Nikola Jokić
player_id = 203999
season = "2025-26"

# Veriyi çek
log = playergamelog.PlayerGameLog(player_id=player_id, season=season)
df = log.get_data_frames()[0]

# İlgili kolonları filtrele
df_selected = df[["GAME_DATE", "MATCHUP", "MIN", "PTS", "REB", "AST", "FG_PCT", "FG3_PCT", "FT_PCT"]]

# CSV dosyası olarak kaydet
csv_path = DATA_RAW / f"player_{player_id}_gamelog_2025_26.csv"
df_selected.to_csv(csv_path, index=False)

print(f"✓ Kaydedildi: {csv_path.resolve()}")
```
