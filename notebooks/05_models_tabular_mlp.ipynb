```python
# Virtual environment'ın Python'unu kullan
import os
from pathlib import Path

project_root = Path().absolute()
if project_root.name == "notebooks":
    project_root = project_root.parent

venv_python = project_root / "venv" / "Scripts" / "python.exe"
if venv_python.exists():
    python_cmd = str(venv_python)
    print(f"✅ Virtual environment Python kullanılıyor: {python_cmd}")
else:
    python_cmd = "python"
    print("⚠️ Virtual environment bulunamadı, sistem Python kullanılıyor")

# Model eğitimini çalıştır
os.system(f'"{python_cmd}" -m src.models.mlp')
```

```python
import json, pandas as pd, os

pred_csv_path = "data_processed/predictions_mlp.csv"
metrics_json_path = "reports/metrics/model_results.json"

print("pred csv:", os.path.exists(pred_csv_path))
print("metrics json:", os.path.exists(metrics_json_path))

# Predictions dosyasını kontrol et
if os.path.exists(pred_csv_path):
    df = pd.read_csv(pred_csv_path)
    print("\nPredictions CSV sütunları:", df.columns.tolist())
    print("\nİlk 5 satır:")
    print(df.head())
else:
    print("\n⚠️ predictions_mlp.csv bulunamadı!")
    print("Önce Cell 2'yi çalıştırın: !python -m src.models.mlp")

# Metrics JSON'u kontrol et
if os.path.exists(metrics_json_path):
    d = json.load(open(metrics_json_path, "r", encoding="utf-8"))
    print("\nModel sonuçları:")
    print("  - mlp_classifier:", "mlp_classifier" in d)
    print("  - mlp_regressor:", "mlp_regressor" in d)
    if "mlp_classifier" in d:
        print("  - Test AUC:", d["mlp_classifier"]["test"]["roc_auc"])
else:
    print("\n⚠️ model_results.json bulunamadı!")
    print("Önce Cell 2'yi çalıştırın: !python -m src.models.mlp")
```
