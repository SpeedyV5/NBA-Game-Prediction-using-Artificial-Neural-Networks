```python
import os
import sys
from pathlib import Path

# ================== SETUP ==================
# Notebook'un proje root'undan çalışması için
project_root = Path().absolute()
if project_root.name == "notebooks":
    project_root = project_root.parent
os.chdir(project_root)

# Virtual environment'ı kullan (eğer varsa)
venv_python = project_root / "venv" / "Scripts" / "python.exe"
if venv_python.exists():
    # Virtual environment'ın site-packages'ını path'e ekle
    venv_site_packages = project_root / "venv" / "Lib" / "site-packages"
    if str(venv_site_packages) not in sys.path:
        sys.path.insert(0, str(venv_site_packages))
    print(f"✅ Virtual environment kullanılıyor: {venv_python}")
else:
    print("⚠️ Virtual environment bulunamadı, sistem Python kullanılıyor")

print("CWD:", os.getcwd())
print("Project root:", project_root)
print("Python executable:", sys.executable)
```
